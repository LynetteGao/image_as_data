# real collection begins, may take a while

```python
transcript_list, unretrievable_videos = YouTubeTranscriptApi.get_transcripts(video_ids, continue_after_error=True,languages = ['en'])

captions = []
counter = 0
    
for video_id in video_ids:
    print(counter)
    print(video_id)
    if video_id in transcript_list.keys():

        srt = transcript_list.get(video_id)

        text_list = []
        for i in srt:
            text_list.append(i['text'])

        text = ' '.join(text_list)
        captions.append(text)
    else:
        captions.append('CaptionUnavailable')
    counter += 1
```

# save the transcript into json file

```python
'''
save the transcript into json file, as a list of dictionaries with form {video_id: transcript}

'''
all_captions = {}
all_captions['transcripts'] = []
for i in range(0,len(captions)):
    key = video_ids[i]
    all_captions['transcripts'].append({key: captions[i]})
    
with open('test.json','w') as f: # change json name if you want
    json.dump(all_captions,f)
```
