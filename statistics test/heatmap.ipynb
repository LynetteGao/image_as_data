```python
import pandas as pd
from mlxtend.plotting import heatmap
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('handlabel_feature.csv')
```

```python
df.columns
```

```python
X = df[df.columns[25:45]]
X.shape
```

```python
X.columns = ['rgbR variance',
 'rgbG variance',
 'rgbB variance',
 'hue variance',
 'saturation variance',
 'value variance',
 'brightness variance',
 'brightness_sd variance',
 'contrast variance',
 'colorful variance',
 'rgbR median',
 'rgbG median',
 'rgbB median',
 'hue median',
 'saturation median',
 'value median',
 'brightness median',
 'brightness_sd median',
 'contrast median',
 'colorful median']
```

```python

cols = ['rgbR variance',
 'rgbG variance',
 'rgbB variance',
 'hue variance',
 'saturation variance',
 'value variance',
 'brightness variance',
 'brightness_sd variance',
 'contrast variance',
 'colorful variance',
 'rgbR median',
 'rgbG median',
 'rgbB median',
 'hue median',
 'saturation median',
 'value median',
 'brightness median',
 'brightness_sd median',
 'contrast median',
 'colorful median']
# plt.figure(figsize=(150,100))
cm = np.corrcoef(X[cols].values.T)
#hm = heatmap(cm, column_names=cols, figsize=(15,15),row_names=cols,cell_values=False)
fig, ax = heatmap(cm, column_names=cols, figsize=(15,15),row_names=cols,cell_values=False)
plt.setp(ax.xaxis.get_majorticklabels(), rotation=45, ha="right", rotation_mode="anchor") 
plt.savefig('m4_corr_novalue2.pdf')  
plt.show()

```
