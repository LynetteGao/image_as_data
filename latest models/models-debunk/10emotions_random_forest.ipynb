```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
big = pd.read_excel("relevant_youtube_dataset.xlsx")
```

```python
big.shape
```

```python
small = pd.read_csv("handlabel.csv")
```

```python
small.shape
```

```python
df = pd.merge(left=small, right=big, left_on='video_id', right_on='video_id')
```

```python
t = df.drop_duplicates(subset=['video_id'])
```

```python
t
```

```python
t.columns
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
t['attitude'] = le.fit_transform(t['attitude'])
```

```python
t.columns.tolist()[45:55]
```

```python
X = t[t.columns[45:55]]
X
```

```python
y = t['attitude']
y
```

```python
from sklearn.metrics import average_precision_score
from sklearn.metrics import recall_score
```

```python
precision = []
recall = []
```

```python
from sklearn.model_selection import cross_validate
from sklearn import datasets
from sklearn.ensemble import RandomForestClassifier

for i in range(10):
    forest = RandomForestClassifier(n_estimators=500,
                                 max_features=None,max_depth=None,min_samples_split=2)
    cv = cross_validate(forest, X, y, cv=10,scoring = ['precision','recall'])
    precision.append(cv['test_precision'].mean())
    recall.append(cv['test_recall'].mean())


```

```python

precision



```

```python
recall
```

```python
import numpy as np
import scipy.stats as st


print(np.std(precision))
print(np.mean(precision))
print(st.t.interval(0.95, len(precision)-1, loc=np.mean(precision), scale=st.sem(precision)))



```

```python

print(np.std(recall))
print(np.mean(recall))
print(st.t.interval(0.95, len(recall)-1, loc=np.mean(recall), scale=st.sem(recall)))




```

```python
import joblib
joblib.dump(forest, "random_forest_10_emotions.joblib")
```

```python
import eli5
from eli5.sklearn import PermutationImportance
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2, shuffle = True, random_state = 123, stratify = y)
model = forest.fit(X_train,y_train)
perm = PermutationImportance(model).fit(X_test, y_test)
eli5.show_weights(perm, feature_names=X_test.columns.tolist())
```
