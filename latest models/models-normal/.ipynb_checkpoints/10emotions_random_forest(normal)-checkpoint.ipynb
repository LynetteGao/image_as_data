```python
t =  pd.read_csv("10emotion_normal.csv")
```

```python
t = t.drop_duplicates(subset=['videoid'])
```

```python
t
```

```python
t.columns
```

```python
t = t[t['conspiracy'] != 0]
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
t['conspiracy'] = le.fit_transform(t['conspiracy'])
```

```python
t
```

```python
t.columns.tolist()[2:12]
```

```python
X = t[t.columns[2:12]]
X
```

```python
y = t['conspiracy']
y
```

```python
from sklearn.metrics import average_precision_score
from sklearn.metrics import recall_score
```

```python
precision = []
recall = []
```

```python
from sklearn.model_selection import cross_validate
from sklearn import datasets
from sklearn.ensemble import RandomForestClassifier
from sklearn.utils import shuffle

for i in range(10):
    t = shuffle(t)
    X = t[t.columns[2:12]]
    y = t['conspiracy']
    forest = RandomForestClassifier(n_estimators=500,
                                 max_features=None,max_depth=None,min_samples_split=2)
    cv = cross_validate(forest, X, y, cv=10,scoring = ['precision','recall'])
    precision.append(cv['test_precision'].mean())
    print(cv['test_precision'])
    recall.append(cv['test_recall'].mean())


```

```python

precision



```

```python
recall
```

```python
import numpy as np
import scipy.stats as st


print(np.std(precision))
print(np.mean(precision))
print(st.t.interval(0.95, len(precision)-1, loc=np.mean(precision), scale=st.sem(precision)))



```

```python

print(np.std(recall))
print(np.mean(recall))
print(st.t.interval(0.95, len(recall)-1, loc=np.mean(recall), scale=st.sem(recall)))




```

```python
import joblib
joblib.dump(forest, "random_forest_10_emotions_normal.joblib")
```
